# Frameworks

Neste ponto, você já deve ter observado que existe uma enorme quantidade de recursos que podemos lançar mão para criar nossos projetos de inteligência artificial.  

As próximas perguntas que devem ter lhe ocorrido são...  
- Ok, mas por onde começar?  
- Como devo conduzir o projeto para ter sucesso?  
- Quando devo parar?  
- E se não der certo, o que fazer?  

E outras tantas.  

Saiba que você não é o primeiro e nem a única a ser assaltada por estas dúvidas. Ainda bem, não?  

Afinal, algúem já deve ter tentado solucionar estes problemas.  

E, a ótima notícia é que SIM! Existem várias iniciativas no sentido de se organizar a forma como você deve conduzir o seu projeto de ciência de dados e posteriormente levar para produção o resultado do seu trabalho.

Tal qual a maior parte das atividades de desenvolvimento do mundo digital, para se obter resultados consistentes, precisamos de uma metodologia que oriente os passos que devemos seguir.  

Para tanto, vamos estudar o modelo **CRISP-DM (Cross-Industry Standard Process for Data Mining)** e entender como ele pode ser benéfico para a condução das nossas iniciativas de uso de inteligência artificial.  


Vale ressaltar que o modelo CRISP-DM foi originalmente concebido para o desenvolvimento de projetos de mineração de dados, mas com o avanço das iniciativas de inteligência artificial, especialmente de machine learning, observou-se que com pequenas adaptações o modelo se apresenta também como uma ótima ferramenta para estes projetos.

## CONTEXTUALIZAÇÃO

Para compreendermos corretamente a proposta do modelo CRISP-DM, precisamos incialmente entender o conceito que está relacionado às metodologias cíclicas.  

Para isto, vamos dar um passo atrás e entender o desenvolvimento destas metodologias.

Segundo o PMBOK®, “Projeto é um esforço temporário empreendido para criar um produto, serviço ou resultado único.” (PMI, 2021).  

Originalmente, os projetos eram desenhados como uma cadeia contínua de atividades, com início e fim claramente determinados. Tal sucessão de atividades era popularmente conhecida como “modelo em cascata” pois uma atividade normalmente se iniciava apenas após o término de sua antecessora.  

Entretanto, observou-se que muitas das iniciativas de desenvolvimento de projetos não apresentavam bons resultados quando orientadas por metodologias em cascata, pois durante o andamento de suas atividades, inevitavelmente mudanças e adaptações surgiam, as quais não podiam ser absorvidas dentro da rígida estrutura inicialmente definida.  

Em função disto, buscou-se novas abordagens que tornassem os processos mais flexíveis e adaptáveis, tornando-os capazes de incorporar mudanças e melhorias, sem se distanciar dos objetivos traçados.  

<center><img src="./images/Metodologias_Ciclicas_01.png" width="50%"></center>

A solução encontrada foi adaptar metodologias cíclicas, as quais, através de etapas recorrentes, permitem o aprimoramento sucessivo dos resultados, convergindo para os objetivos do projeto. 

<center><img src="./images/Metodologias_Ciclicas_02.png" width="50%"></center>

Interessantemente, o conceito da abordagem cíclica não é novo. A abordagem mais conhecida deste conceito é a metodologia PDCA, também chamada de ciclo de Shewhart ou ciclo de Deming.  

O PDCA teve suas origens a partir do Controle Estatístico de Processos, ferramenta orientada ao controle de qualidade desenvolvida pelo engenheiro Water Shewhart na década de 20 do século passado, o qual propôs um ciclo de melhoria contínua composto por quatro etapas sucessivas – Plan, Do, Check, Act.  

Contudo, foi durante a reconstrução do Japão pós segunda guerra mundial que o PDCA foi popularizado através das ações do consultor de gestão William Edwards Deming, sendo adotado pela indústria japonesa e posteriormente se tornando um dos pilares do Sistema Toyota de Produção conhecido com Kaizen.
 
<center><img src="./images/pdca.png" width="50%"></center>

Como pode ser visto no diagrama acima, o ciclo PDCA não se encerra necessariamente após a primeira iteração. Na verdade, admite-se várias iterações onde a cada novo ciclo, as lições aprendidas dos ciclos anteriores são incorporadas o que tende a convergir para uma solução ótima.
Apesar do ciclo PCDA estar intimamente ligado aos processos de garantia da qualidade, a abordagem cíclica foi adotada por diversos modelos de trabalho.
Destes, podemos citar o método Kaisen, o Lean Thinking e o Six Sigma com o ciclo DMAIC (Define, Measure, Analyze, Improve, Control) no ambiente industrial; e o Extreme Programming e o método SCRUM no mundo dos projetos de desenvolvimento de software.

## O CRISP-DM

 O modelo CRISP-DM foi inicialmente concebido no final de 1996 por três veteranos (DaimlerChrysler, SPSS, NCR) do ainda imaturo mercado de data mining. A ideia foi criar um modelo de processo padronizado, bem definido, não proprietário e disponível gratuitamente.  

A partir desta iniciativa, foi criado um consórcio denominado “CRISP-DM Special Interest Group”, sendo que em meados de 1999, a versão 1.0 do modelo foi lançada.  

### O modelo de referência CRISP-DM  

O modelo de referência 1.0 do CRISP-DM fornece uma visão abrangente e bem definida do ciclo de vida de um projeto de mineração de dados.  
 
O ciclo de vida é dividido em 6 fases distintas. Como pode ser observado acima, as fases do CRISP-DM são Entendimento do Negócio, Entendimento dos Dados, Preparação dos Dados, Modelagem, Avaliação e Implementação. Na mesma figura também pode-se observar a direção das setas que nos indicam a sequência natural dos ciclos, mas também temos as setas nos indicando o movimento reverso. Isto quer dizer que a sequência das fases não é rígida, sendo que é possível avançar ou retroceder medida da necessidade observada durante o desenvolvimento do projeto.  

Além disso, o projeto pode ser composto de quantas iterações se façam necessárias até que se obtenha o resultado esperado ou mesmo que se comprove a impossibilidade de obtê-lo.  

As primeiras três etapas do modelo estão relacionadas à compreensão e coleta dos dados que serão necessários para servir de subsídios para as etapas seguintes. 

Já as três etapas seguintes visam a criação de um modelo que resolva o problema de negócios identificado na primeira fase, e se possível, implementá-lo.  

O vídeo [CRISP-DM: a melhor metodologia para projetos de Data Science](https://www.youtube.com/watch?v=mhPATD8S6D0) apresenta uma visão rápida do modelo CRISP-DM.

Ou, se você quiser ter uma visão mais abrangente, explore a [playlist](https://www.youtube.com/playlist?list=PLbmt8d_ueDMVEo273jJYu4ssrvhZkDsX3) do prof. Sandeco.

Com tudo isto posto, vamos descrever cada uma das fases para que possamos obter a completa compreensão do funcionamento modelo.  

### Business Understanding
A etapa de Entendimento do Negócio representa a primeira das atividades de um ciclo CRISP-DM e, de certa forma, podemos dizer que é a mais importante delas, pois a execução inadequada desta etapa pode nos comprometer o sucesso de todas as etapas seguintes. É nesta etapa que vamos procurar identificar o problema que se deseja resolver, sob a ótica de valor agregado aos negócios. É neste momento que será definido o objetivo do projeto, alinhadas as expectativas dos stakeholders e, principalmente, definidos os critérios de sucesso do projeto.  

É importante que se incluam as metas que devem ser alcançadas, os requisitos mínimos que devem ser respeitados, as áreas envolvidas, e todas as hipóteses que se desejam testar.  

Outro aspecto importante é estabelecer **critérios objetivos** ao invés de **critérios vagos**.  

Via de regra, critérios objetivos podem ser respondidos por um “nome” ou por um “número” ao passo que critérios vagos não admitem uma resposta, mas sim um conceito.  

Por exemplo, quando alguém apresenta uma pergunta do tipo “Como posso aumentar o meu lucro”, estamos falando de algo subjetivo, vago, que admite diversas respostas. Por outro lado, quando alguém faz a seguinte pergunta: “Quais produtos devo colocar em promoção para aumentar em 10% as compras cruzadas”, estamos estabelecendo critérios claros que podem ser medidos de forma objetiva e, desta forma, nos indicar se estamos obtendo sucesso e o quanto ainda falta para alcança-lo.  


### Data Understanding
A etapa de Entendimento dos Dados se inicia com a coleta preliminar de dados e segue com as atividades para que se familiarizar com os dados, identificando suas características e eventuais problemas de qualidade neles contidos.  

Para isso devemos realizar vários questionamentos sobre os dados e sua natureza como: 
- “Os dados estão estruturados ou não?   
- Iremos acessar dados contidos em algum banco de dados da empresa ou utilizaremos outras fontes como arquivos .csv ou planilhas de cálculo?  
- Teremos acesso aos dados ou existe alguma restrição de governança que nos impede de acessá-los? Temos que observar alguma regra específica relacionada a privacidade e proteção de dados?  

E assim por diante.  

Outro aspecto importante é a limpeza e pré-processamento dos dados.  
É muito comum encontrar bases contendo valores inválidos ou organizados de uma forma que não nos permite avançar para as fases seguintes.  

Também não é incomum que tenhamos que “_enriquecer_” a base de dados a partir de fontes externas, as quais da mesma forma podem não estar adequadas para o consumo das etapas seguintes. 

E finalmente, é nesta etapa ocorre o que chamamos de **EDA (exploratory data analysis – análise exploratória dos dados)**. Momento em que realizamos várias análises estatísticas e obtemos vários insights sobre os dados disponíveis.  

Explore a aula de [análise exploratória de dados](https://www.youtube.com/watch?v=BHk5fdfDNZU) do curso de ciência de dados e machine learning do Instituto de Ciências Matemáticas e de Computação da USP.  

No [Kaggle](https://www.kaggle.com/code/imoore/intro-to-exploratory-data-analysis-eda-in-python) também temos um curso introdutório sobre o assunto. Vale dar uma olhada.

O documento [10 Dicas para uma Boa Apresentação](./docs/10-Dicas-para-uma-Boa-Apresentacao.pdf) nos trás algumas informações adicionais sobre a forma de apresentar suas descobertas.


### Data Preparation
Após coletar os dados e realizar as primeiras análises, chega o momento de organizar os dados de modo que possam ser consumidos pelas etapas seguintes.  

E isto porque, nem sempre os dados estão devidamente organizados, posto que estamos nos referindo a tecnologias diferentes daquelas que tradicionalmente encontramos nos sistemas e processos corporativos.  

Neste momento algumas questões acabam surgindo:  
- Como devemos tratar os dados nulos?  
- Quais atributos (features) usaremos?  
- Precisaremos criar atributos derivados?  
- Será necessário fazer a transformação de algum atributo (ex. categórico para numérico)?  
- Como será o formato final dos dados?  
- Precisaremos unir conjuntos distintos?  

E assim por diante.

Como nesta etapa vamos trabalhar e transformar praticamente toda a massa de dados disponível, esta tende a ser a etapa mais demorada do processo.  

Mas, vale o investimento pois como já sabemos, os modelos de machine learning precisam de dados para gerar seus resultados. E obviamente, se os insumos entregues para os modelos não estiverem adequados, não obteremos bons resultados. Aliás no mundo do machine learning é comum ouvir-se a expressão: “Garbage in – Garbage out”, ou seja, "se entra lixo, sai lixo".

### Modeling  

Nesta fase é que aplicamos as técnicas de machine learning, métodos estatísticos e outros recursos para endereçarmos a solução do problema de negócios definido na primeira etapa do modelo.  

Este é o momento de criar os modelos, efetuar ajustes e coletar as métricas visando testar as hipóteses levantadas e atingir os critérios de sucesso previamente definidos.  

Vale lembrar que apesar dos esforços dispendidos nas etapas anteriores, eventualmente os dados não se apresentam da forma adequada para serem consumidos pelos algoritmos selecionados. Caso isto aconteça, é o momento de executar novamente a segunda e/ou terceira etapa.  

Também faz parte desta etapa realizar ajustes nos hiper parâmetros dos modelos para se conseguir melhorar as métricas obtidas.  

Então, é muito comum que aconteçam várias iterações entre a segunda, terceira e quarta etapas para se obter os melhores resultados possíveis, observando-se critérios de sucesso técnicos.  

### Evaluation

Após termos criado vários modelos e realizado os ajustes finos para obter as métricas que estejam de acordo com os critérios de aceite técnicos, chega o momento de verificar se os resultados obtidos também estão aderentes aos critérios de sucesso de negócios que foram definidos na etapa de business understanding.  

Isto irá nos guiar para selecionar os modelos de machine learning que melhor se adequam à solução almejada e também irá nos indicar eventuais descobertas que enriqueçam o ambiente de negócios.  

A propósito, algumas perguntas podem nortear nossas ações neste momento:  

- Os resultados obtidos estão organizados de forma que possam ser facilmente apresentados?  
- Foram detectadas descobertas novas ou exclusivas que devem ser destacadas?  
- É possível organizar de forma hierárquica os modelos e as descobertas de acordo com sua aplicabilidade às metas dos negócios?  
- Como os resultados respondem às metas de negócios da organização?  
- Quais outras perguntas os resultados levantaram?  
- É possível expressar essas perguntas em termos de negócios?  

E, finalmente, deve-se produzir um relatório final que irá nortear as decisões de negócios relacionadas aos objetivos alcançados.

### Deployment

Com os resultados documentados, as métricas coletadas e os critérios de sucesso atingidos, chega o momento em que os modelos devem ser entregues para serem utilizados em ambiente de produção e este é o objetivo da fase de implantação.  

Entretanto, apesar de parecer singela, muitos dos projetos de machine learning falham exatamente neste ponto. E, não é à toa que uma especialidade surgiu para endereçar este problema – o DevOps.  

Não é objetivo deste curso entrar nos detalhes de como realizar a implementação final dos modelos, mas é importante que isto aconteça concomitantemente com as práticas de governança e segurança da corporação.
